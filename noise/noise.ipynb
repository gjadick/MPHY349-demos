```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 100
M = np.zeros([n,n])  # background
mask = M.copy().astype(bool)

sz = 5
mask[20:20+sz, 30:30+sz] = True

plt.imshow(mask)
plt.show()
```

```python
def SNR(u1,s1,u2,s2):
    return (u2-u1)/np.sqrt(0.5*s1**2 + 0.5*s2**2)
```

```python
fig,ax=plt.subplots(1,3,dpi=150)
for axi in ax:
    axi.axis('off')
ax[0].imshow(mask)
fsz=6

u_b = 100.0
s_b = 2.0
u_s = 110.0
s_s = 1.0
snr = SNR(u_b,s_b,u_s,s_s)
M_noisy = np.random.normal(u_b, s_b, n*n).reshape([n,n])
signal_noisy = np.random.normal(u_s, np.sqrt(s_s**2 + s_b**2), sz*sz).reshape([sz,sz])
M_signal_noisy = M_noisy.copy()
M_signal_noisy[20:20+sz, 30:30+sz] = signal_noisy
ax[1].imshow(M_signal_noisy)
ax[1].set_title(f'{u_b:.0f}$\pm${s_b:.0f} vs. {u_s:.0f}$\pm${s_s:.0f} (SNR={snr:.1f})', fontsize=fsz)

u_b = 100.0
s_b = 10.0
u_s = 110.0
s_s = 1.0
snr = SNR(u_b,s_b,u_s,s_s)
M_noisy = np.random.normal(u_b, s_b, n*n).reshape([n,n])
signal_noisy = np.random.normal(u_s, np.sqrt(s_s**2 + s_b**2), sz*sz).reshape([sz,sz])
M_signal_noisy = M_noisy.copy()
M_signal_noisy[20:20+sz, 30:30+sz] = signal_noisy
ax[2].imshow(M_signal_noisy)
ax[2].set_title(f'{u_b:.0f}$\pm${s_b:.0f} vs. {u_s:.0f}$\pm${s_s:.0f} (SNR={snr:.1f})', fontsize=fsz)

fig.tight_layout()
plt.savefig('examples/snr.png')
plt.show()
```
